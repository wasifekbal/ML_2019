```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
match = pd.read_csv('matches.csv')
delivery = pd.read_csv('deliveries.csv')
```

# 1. Pie Charts

```python
def btmn_rcrd(batsman):
    vk = delivery[delivery['batsman']==batsman]
    y = vk['batsman_runs'].value_counts().sort_values()
    plt.pie(y,labels=y.index,autopct='%0.2f%%')
    return plt.show()


y = input('Enter the Player\'s name :')  
btmn_rcrd(y)
'''
vk = delivery[delivery['batsman']=='V Kholi']
y = vk['batsman_runs'].value_counts().sort_values()
plt.pie(y,labels=y.index,autopct='%0.2f%%')
plt.show()'''
```

# 2. Line Chart

```python
def comp_plrs(player1,player2):
    mega = match.merge(delivery,left_on='id',right_on='match_id')
    batsman_data = mega[mega['batsman']==player1]
    batsman_data1 = mega[mega['batsman']==player2]
    record = batsman_data.groupby('season').sum()['batsman_runs']
    record1 = batsman_data1.groupby('season').sum()['batsman_runs']
    plt.plot(record.index, record.values,color='red', marker='*',label=player1)
    plt.plot(record1.index, record1.values,color='blue', marker='+',label=player2)
    plt.xlabel('season')
    plt.ylabel('Runs Scored')
    plt.title('Batsman Season by season performance')
    plt.grid()
    plt.legend()

name1 = input('Enter the first Player\'s name :')
name2 = input('Enter the second Player\'s name :')
comp_plrs(name1,name2)
```
