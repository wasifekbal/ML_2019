```python
import seaborn as sns
tips=sns.load_dataset('tips')
tips.head()
```

```python
'''
Distribution graphs :
 1. distplot.
 2. jointplot.
 3. pairplot.
 4. rugplot.
 
Categorical Plots :
 1. barplot.
 2. countplot.
 3. boxplot.
 4. violinplot.

matrix plot :
 1. Heatmap
 
'''
```

## 1. distplot

```python
sns.distplot(tips['total_bill'],kde=False)
```

## 2. Jointplot

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex',color='brown')
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde',color='black')
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg',color='black')
sns.jointplot(x='total_bill', y='tip', data=tips, color='black')
```

# 3. Pairplot

```python
sns.pairplot(tips, hue='sex')
```

# 4. Rugplot

```python
sns.rugplot(tips['total_bill'])
```

# Categorical Plots :

# 1. Barplot :

```python
sns.barplot(x='day',y='tip',data=tips, hue='sex')
```

```python
sns.barplot(x='day',y='tip',data=tips, hue='smoker')
tips
```

## 2. Countplot :

```python
import numpy as np
sns.countplot(tips['smoker'], data=tips, hue='sex',)
```

## 3. Box Plot

```python
#sns.boxplot('total_bill',data=tips, hue='sex')
sns.boxplot(x=tips['sex'],y=tips['tip'])
```

## 4. Violinplot :

```python
sns.violinplot(x=tips['day'],y=tips['tip'],)
```
