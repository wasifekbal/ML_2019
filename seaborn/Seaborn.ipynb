```python
import seaborn as sns
tips=sns.load_dataset('tips')
tips.head()
```

```python
'''
Distribution graphs :
 1. distplot.
 2. jointplot.
 3. pairplot.
 4. rugplot.
 
Categorical Plots :
 1. barplot.
 2. countplot.
 3. boxplot.
 4. violinplot.

matrix plot :
 1. Heatmap
 
'''
```

## 1. distplot

```python
sns.distplot(tips['total_bill'],kde=False)
```

## 2. Jointplot

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex',color='brown')
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde',color='black')
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg',color='black')
sns.jointplot(x='total_bill', y='tip', data=tips, color='black')
```

# 3. Pairplot

```python
sns.pairplot(tips, hue='sex')
```

# 4. Rugplot

```python
sns.rugplot(tips['total_bill'])
```
